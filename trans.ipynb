```python
import csv

def txt_to_csv(input_file, output_file, delimiter='\t', encoding='utf-8'):
    with open(input_file, 'r', encoding=encoding) as txt_file:
        # 从TXT文件读取数据
        txt_data = [line.strip().split(delimiter) for line in txt_file.readlines()]

    with open(output_file, 'w', newline='', encoding='utf-16') as csv_file:
        # 使用CSV模块写入CSV文件
        csv_writer = csv.writer(csv_file, delimiter=delimiter)
        csv_writer.writerows(txt_data)

if __name__ == "__main__":
    # 替换这里的文件路径
    input_txt_file = r"E:\桌面\military_triples_from_baiketriples.txt"
    output_csv_file = 'E://桌面/3.csv'

    txt_to_csv(input_txt_file, output_csv_file)
```

### 汽车飞机三元组提取工作

#### txt转为csv

```python
import csv
import os

# 以空格为分割符
def txt_to_csv(txt_file_path, csv_file_path):
    # 读取txt文件内容
    with open(txt_file_path, 'r') as txt_file:
        txt_data = txt_file.readlines()

    # 写入csv文件内容
    with open(csv_file_path, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        for line in txt_data:
            # 去除每行末尾的换行符，并按空格分割
            row_data = line.strip().split(' ')
            # 写入csv文件
            writer.writerow(row_data)




# 输入需要处理的txt文件所在的文件夹
input_folder = r"E:\桌面\飞机汽车三元组构建工作\ship_label"

# 输出csv文件的路径
output_folder = r"E:\桌面\飞机汽车三元组构建工作\ship_label_csv"

# 确保输出文件夹存在，如果不存在则创建
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    try:
        # 获取文件名（不含扩展名）
        file_name, _ = os.path.splitext(filename)
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, file_name + ".csv")
        txt_to_csv(input_file_path, output_file_path)
    except Exception as e:
        print(f"遇到错误，文件名：{filename}，错误信息：{e}")

```

```python
import os
import csv

def extract_column(csv_file, column_index):
    extracted_data = set()  # 创建一个空集合
    with open(csv_file, 'r', newline='', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            if len(row) > column_index:
                extracted_data.add(row[column_index])  # 将元素添加到集合中
    return extracted_data  # 直接返回集合，不再转换为列表

input_folder_csv = r"E:\桌面\飞机汽车三元组构建工作\ship_label_csv"
column_index = 8  # 要提取的列的索引（从0开始）

for filename in os.listdir(input_folder_csv):
    try:
        # 提取不包含扩展名的文件名部分
        file_name = os.path.splitext(filename)[0]
        input_csvfile_path = os.path.join(input_folder_csv, filename)
        extracted_column = extract_column(input_csvfile_path, column_index)
        # 打印文件名（不含扩展名）和去重元素
        print(f"文件名: {file_name}, 去重元素: {extracted_column}")
    except Exception as e:
        print(f"遇到错误，文件名：{filename}，错误信息：{e}")
```

#### 保存为txt文件

```python
import os
import csv

def extract_column(csv_file, column_index):
    extracted_data = set()  # 创建一个空集合
    with open(csv_file, 'r', newline='', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            if len(row) > column_index:
                extracted_data.add(row[column_index])  # 将元素添加到集合中
    return extracted_data  # 直接返回集合

input_folder_csv = r"E:\桌面\飞机汽车三元组构建工作\ship_label_csv"
column_index = 8  # 要提取的列的索引（从0开始）

output_file = "output1.txt"

with open(output_file, 'w', encoding='utf-8') as f:
    for filename in os.listdir(input_folder_csv):
        try:
            # 提取不包含扩展名的文件名部分
            file_name = os.path.splitext(filename)[0]
            input_csvfile_path = os.path.join(input_folder_csv, filename)
            extracted_column = extract_column(input_csvfile_path, column_index)
            # 将文件名、contain和去重元素写入到文本文件中
            for item in extracted_column:
                f.write(f"{file_name} contain {item}\n")
        except Exception as e:
            print(f"遇到错误，文件名：{filename}，错误信息：{e}")
```
